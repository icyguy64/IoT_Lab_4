```python
import boto3
client = boto3.client('iotanalytics')
import pandas as pd

dataset_url = client.get_dataset_content(datasetName =
"iotlab4_dataset")['entries'][0]['dataURI']
data = pd.read_csv(dataset_url)
```

```python
data.describe()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
data.head()
```

```python
plt.plot(data["vehicle_co2"])
```

```python
plt.figure(figsize=(10,6), tight_layout=True)
#plotting
plt.plot(data["vehicle_co2"], 'o-', linewidth=2)
#customization
#plt.xticks([2017, 2018, 2019, 2020, 2021])
plt.xlabel('Emissions (CO2)')
plt.ylabel('Time')
plt.title('Emissions of CO2 over time')
#plt.legend(title='Players', title_fontsize = 13, labels=['L. Messi', 'Cristiano Ronaldo', 'K. De Bruyne', 'V. van Dijk', 'K. Mbappé'])
plt.show()
```

```python
plt.figure(figsize=(10,6), tight_layout=True)
ax = sns.scatterplot(data=data["vehicle_co2"], palette='Set2', s=60)
ax.set(xlabel='Time', ylabel='Emissions (CO2))')
ax.legend(title='CO2 Emissions', title_fontsize = 12) 
plt.show()
```

```python
plt.figure(figsize=(10,6), tight_layout=True)
# matplotlib
bins = [1000,2000,3000,4000,5000,6000,7000,8000,9000]


plt.hist(data["vehicle_co2"], bins=bins, color=sns.color_palette('Set2')[2], linewidth=2, edgecolor='black',orientation='horizontal')
plt.title('Histogram of vehicle CO2 emission')
plt.ylabel('Numbers')
plt.xlabel('CO2 emission')
```

```python
# seaborn
bins = [1000,2000,3000,4000,5000,6000,7000,8000,9000]

ax = sns.histplot(data=data["vehicle_co2"], bins=bins, color=sns.color_palette('Set2')[2], linewidth=2)
ax.set(title='Histogram of vehicle CO2 emission', xlabel='CO2_emissions bin', ylabel='Numbers')
plt.show()

```
